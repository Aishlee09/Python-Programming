```python
# ATM Simulator
print('ðŸ˜Š WELCOME TO OUR SERVICE ðŸ˜Š')
print()
print('Insert your card in the ATM slot')

user_pin = int(input("\nEnter your PIN: "))
ini_balance = 20000
default_pin = 1234
while True:
    # check validation
    if user_pin == default_pin:
        print(''' SELECT THE SERVICE
        Cash Withdrawal-1
        Balance Check-2
        Deposit-3
        Pin Change-4
        ''')
        # control structure
        option = int(input("Enter your option: "))
        if option == 1:
            wd_amount = int(input('\nEnter the amount to withdaw: '))
            if wd_amount<=ini_balance:
                print('Your transaction is in process.....')
                print('\nCollect the money and Take your card.')
                print('Transaction Successful.')
                ch = input('\nDo you want to print the balance? ')
                ini_balance = ini_balance-wd_amount
                if ch == 'yes':
                    print("Your balance is: ", ini_balance)
                else:
                    print("THANK YOU :)")
            else:
                print("Insufficient Amount.")
        elif option == 2:
            print("Your balance is: ", ini_balance)
        elif option == 3:
            d_amount = int(input("Enter the amount to deposit: "))
            ini_balance+=d_amount
            print('Your transaction is in process.....')
            print('Transaction Successful.')
            ch = input('\nDo you want to print the balance? ')
            if ch == 'yes':
                print("Your balance is: ", ini_balance)
            else:
                print("THANK YOU :)")
        elif option == 4:
            org_pin = int(input("Enter your original PIN: "))
            if org_pin == default_pin:
                new_pin = int(input("Enter your new PIN: "))
                conf_pin = int(input("Enter your confirm PIN: "))
                if new_pin == conf_pin:
                    #default_pin = new_pin
                    print("PIN changed successfully.")
                else:
                    print("Your PIN doesn't match.")
            else:
                print("Invalid PIN. Enter the correct PIN.")
    choice = input("Do you want to use our services again? (yes/no): ")
    if choice == 'yes':
        continue
    else:
        print("\n Thanks for using our services.")
        break
```
