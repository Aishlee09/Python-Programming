```python
# With grid as geometry manager
import tkinter
from tkinter import *

# window
root = tkinter.Tk()  # Tk()  is used to create a class
root.title('First window')
root.geometry('700x300+300+150')
root.configure(bg = 'cyan')
root.resizable(0,0)

l1 = Label(master = root, text = 'Username', bg = 'blue', width = 10, height = 2, bd = 5, 
          font = 'timesnewroman 14 bold', fg = 'white', relief = 'groove')
l1.grid(row = 0, column = 0)

l2 = Label(master = root, text = 'Password', bg = 'blue', width = 10, height = 2, bd = 5, 
          font = 'timesnewroman 14 bold', fg = 'white', relief = 'groove')
l2.grid(row = 1, column = 0)

# function
def test():
    username = StringVar()
    password = StringVar()
   
    # get()
    username = e1.get()
    password = e2.get()
 
    # database connectivity
    import mysql.connector as co
    # open connection
    connection = co.connect(host = 'localhost', user = 'root', passwd = 'aaabb123',
                         database = 'batch2_2022', auth_plugin = 'mysql_native_password')
    get_data = 'select * from test5_data where na=%s'
   
    cur = connection.cursor()
    cur.execute(get_data, (username,))
    result = cur.fetchall()
    print(result)
    print('Success')
    # close the connection
    connection.commit()   
   
    if username == result[0][0] or password == result[0][4]:
        print('Login Successful')
    else:
        print('Invalid username or password')

# button
b1 = Button(master = root, text = 'Submit', bg = 'blue', width = 10, height = 2, bd = 5, font = 'timesnewroman 14 bold', 
            fg = 'white', relief = 'sunken', activebackground = 'green', activeforeground = 'red', command = test)
b1.place(x = 540, y = 130, width = 120, height = 50)

# entry
e1 = Entry(root, width = 30, textvariable = 'username', font = 'timesnewroman 24', fg = 'black', bd = 2, relief = 'ridge')
e1.grid(row = 0, column = 1)
e2 = Entry(root, width = 30, textvariable = 'password', font = 'timesnewroman 24', fg = 'black', bd = 2, relief = 'ridge')
e2.grid(row = 1, column = 1)       

root.mainloop()  # used to return the window (show window)
```
