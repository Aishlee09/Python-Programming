```python
import tkinter as tk
from tkinter import *
import cv2

root = Tk()
root.title("Image GUI")
root.geometry("500x650+0+0")
root.configure(background = "#FF1493")
root.resizable(0,0)
title = Label(root, text = "Reading Images", font = ("Calibri", 30), bg = "#FF1493", fg = 'black')
title.place(x = 120, y = 18)

frame = Frame(root, bd = 10, bg = '#E0FFFF')
frame.place(x = 60, y = 90, width = 380, height = 500)

i = cv2.imread('cricket.png')
img = cv2.resize(i, (300, 300))
    
def b1():
    l1 = Label(root, text = ' Original image is displayed.ðŸ˜Š', bg = '#FF1493', font = 'Arial 15', fg = 'white')
    l1.place(x = 105, y = 605)
    cv2.imshow('Original Image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def b2():
    l2 = Label(root, text = 'Gray image is displayed. ðŸ˜Š   ', bg = '#FF1493', font = 'Arial 16', fg = 'white')
    l2.place(x = 116, y = 605)
    g = cv2.imread('cricket.png', 0)
    gray = cv2.resize(g, (300, 300))
    cv2.imshow('Gray Image', gray)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def b3():
    l3 = Label(root, text = 'Canny image is displayed. ðŸ˜Š  ', bg = '#FF1493', font = 'Arial 16', fg = 'white')
    l3.place(x = 115, y = 605)
    # canny filter
    can = cv2.Canny(img, 20, 100)
    cv2.imshow('Canny Image', can)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def b4():
    l4 = Label(root, text = ' Blur image is displayed. ðŸ˜Š  ', bg = '#FF1493', font = 'Arial 16', fg = 'white')
    l4.place(x = 115, y = 605)
    # blur filter
    blur = cv2.blur(img, (20, 20))   # kernel matrix (function)
    cv2.imshow('Blur Image', blur)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def b5():
    cv2.destroyAllWindows()
    root.destroy()

b1 = Button(frame, text = 'ORIGINAL Image', fg = 'white', bg = '#1E90FF', relief = 'groove', font = ("Calibri", 17), command = b1)
b1.place(x = 55, y = 30, width = 250, height = 60)

b2 = Button(frame, text = 'GRAY Filter', fg = 'white', bg = '#1E90FF', relief = 'groove', font = ("Calibri", 17), command = b2)
b2.place(x = 55, y = 130, width = 250, height = 60)

b3 = Button(frame, text = 'CANNY Filter', fg = 'white', bg = '#1E90FF', relief = 'groove', font = ("Calibri", 17), command = b3)
b3.place(x = 55, y = 230, width = 250, height = 60)

b4 = Button(frame, text='BLUR Filter', fg = 'white', bg = '#1E90FF', relief = 'groove', font = ("Calibri", 17), command = b4)
b4.place(x = 55, y = 330, width = 250, height = 60)

b5 = Button(frame, text = 'QUIT', fg = 'white', bg = 'black', relief = 'groove', font = ("Calibri", 17), command = b5)
b5.place(x = 105, y = 420, width = 150, height = 40)

root.mainloop()
```
