```python
import tkinter as tk
from tkinter import *
import random
import PIL
from PIL import Image, ImageTk

root = tk.Tk()
root.title("Random Number Guessing Game")
root.configure(bg = '#FFF0F5')
root.geometry('550x600+300+200')
root.resizable(height=False, width=False)

l = Label(root, text='Number Guessing Game', bg='#800000', width=32, height=1,
          font='Arial 19 bold', fg='white')
l.place(x=30, y=25)
l1 = Label(root, text='Try and guess a random number between 1 and 20.\nYou have 8 attempts to guess a number.',
           font='Calibri 14', fg='black', bg='#FFF0F5')
l1.place(x=67, y=172)

# show image
img = ImageTk.PhotoImage(file = r"C:\Users\HP\number.jpg")
img_l=Label(root, image=img, width=465, height=80)
img_l.place(x=40, y=75)

frame = Frame(root, bg = '#00CED1', width=484, height=330)
frame.place(x=32, y=235)
l2 = Label(frame, text='Guess a number', bg='#00CED1', font='Arial 30', fg='white')
l2.place(x=95, y=20)

l3 = Label(frame, bg='#800000', font='Arial 21 bold', fg='white')
l3.place(x=0, y=275)

e1 = Entry(frame, bg='white', font='Arial 35', width=15, fg='red', bd=3, relief='groove')
e1.place(x=45, y=85)

l4 = Label(root, bg='#FFF0F5', font='Arial 14', fg='black')
l4.place(x=100, y=570)

GUESS = 5
win_num = random.randint(1,20)

# function
def check():
    global GUESS
    num = int()
    num = e1.get()
    GUESS-=1
    if int(num)==win_num:
        l3.configure(text='    Congrats! You won the game! üòÉ   ')  
    elif GUESS==0:
        l3.configure(text="You're out of attempts, Try again üëç"  )
        l4.configure(text=f'The number entered by the system is: {win_num}')
        print(f'The number entered by the system is: {win_num}')
    elif int(num)<win_num:
        l3.configure(text='         Your GUESS was small...!        ')
        l4.configure(text="You have "+ str(GUESS) + " attempts remaining! Go Higher!")
    elif int(num)>win_num:
        l3.configure(text='           Your GUESS was big...!          ')
        l4.configure(text="You have "+ str(GUESS) + " attempts remaining! Go Lower!")
    return

def stop():
    root.destroy()
    
b1 = Button(frame, text='Check Guess', bg='yellow', width=12, bd=1, font='Calibri 24', 
            fg='black', relief='groove', command=check)
b1.place(x=70, y=175)

b1 = Button(frame, text='Quit', bg='yellow', width=6, bd=1, font='Calibri 22', 
            fg='black', relief='groove', command=stop)
b1.place(x=300, y=178)

root.mainloop()
```
